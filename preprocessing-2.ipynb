```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.drop_duplicates(subset ="Name", keep = 'last', inplace = True)
```

```python
data.head()
```

```python
data_s=pd.read_csv("anime_with_synopsis.csv")
```

```python
data_s.head(10)
```

```python
data_new = pd.merge(data,data_s, on='MAL_ID')
```

```python
data_new.head()
```

```python
data_new.rename(columns = {'Name_x':'Name'}, inplace = True)
data_new.rename(columns = {'Genres_x':'Genres'}, inplace = True)
data_new.rename(columns = {'Score_x':'rating'}, inplace = True)
```

```python
data_new.head()
```

```python
data_new.drop("Name_y", axis=1, inplace=True)
```

```python
data_new.drop("Score_y", axis=1, inplace=True)
```

```python
data_new.drop("Genres_y", axis=1, inplace=True)
```

```python
data_new.head()
```

```python
data_new.iloc[0].Genres
```

```python
data_new['Genres']=data_new['Genres'].str.split(',')
```

```python
data_new.iloc[0].Genres
```

```python
data_new.head(10)
```

```python
data_new['sypnopsis'] = data_new['sypnopsis'].apply(lambda x:str(x).split())
```

```python
data_new.head()
```

```python
data_new['Studios'] = data_new['Studios'].apply(lambda x:str(x).split())
data_new['Type'] = data_new['Type'].apply(lambda x:str(x).split())
```

```python
data_new.head()
```

```python
data_new['Rating'] = data_new['Rating'].apply(lambda x:str(x).split())
```

```python
data_new['tags'] = data_new['sypnopsis'] + data_new['Genres'] + data_new['Studios'] + data_new['Type'] + data_new['Rating']
```

```python
data_new.head()
```

```python
new = data_new.drop(columns=['sypnopsis','Genres','Studios','Type','Rating'])
#new.head()
```

```python
new.head()
```

```python
new.to_csv('new.csv',index=False)
```
