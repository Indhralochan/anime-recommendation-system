```python
# recommendation will be based on these features only
df = df.loc[:,['MAL_ID','Name', 'Score', 'Genres', 'English name','Type', 'Episodes','Studios','Rating', 'Ranked', 'Popularity']]
```

```python
df.head(10)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df['Genres'] = df['Genres'].replace(np.nan, 'unknown')
df['Rating'] = df['Rating'].replace(np.nan, 'unknown')
df['Ranked'] = df['Ranked'].replace(np.nan, 'unknown')
df['Episodes'] = df['Episodes'].replace(np.nan, 'unknown')
df['Studios'] = df['Studios'].replace(np.nan, 'unknown')
df['Score'] = df['Score'].replace(np.nan, 'unknown')
```

```python
df['Name'] = df['Name'].str.lower()
```
